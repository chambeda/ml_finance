```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(11111111)
```

```python
mean_return = 0.01 * np.random.randn(2)
mean_return
```

```python
rho = 0.01 * np.random.randn()
rho
```

```python
sigmas = np.exp(np.random.randn(2))
sigmas
```

```python
cov = np.diag(sigmas**2)
cov
```

```python
sigma12 = sigmas[0] * sigmas[1] * rho
cov[0,1] = sigma12
cov[1,0] = sigma12
cov
```

```python
def softmax(w):
  a = np.exp(w)
  return a/ a.sum()
```

```python
N = 1000
returns = np.zeros(N)
risks = np.zeros(N)
for i in range(N):
  w = softmax(np.random.randn(2))
  ret = mean_return.dot(w)
  risk = np.sqrt(w.dot(cov).dot(w))
  returns[i] = ret
  risks[i] = risk
```

```python
plt.scatter(risks, returns, alpha=0.1);
plt.xlabel("Risk");
plt.ylabel("Return");
```

```python
N = 1000
returns = np.zeros(N)
risks = np.zeros(N)
for i in range(N):
  x = np.random.random() - 0.5
  w = np.array([x, 1-x]) # short-selling allowed
  ret = mean_return.dot(w)
  risk = np.sqrt(w.dot(cov).dot(w))
  returns[i] = ret
  risks[i] = risk
```

```python
plt.scatter(risks, returns, alpha=0.1);
```
